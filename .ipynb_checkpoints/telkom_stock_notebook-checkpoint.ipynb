```python
import numpy as np 
import pandas as pd 
import math
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import re

from statsmodels.tsa.seasonal import seasonal_decompose

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout

import warnings
warnings.filterwarnings("ignore")

sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
df=pd.read_csv(
    'telkom_stock_data.csv',
    parse_dates=['date'],
    infer_datetime_format=True
).set_index('date').to_period('D')

df.head()
```

# Data Preprocessing

```python
df.shape[0]
```

```python
df.info()
```

### Removing Null Values

```python
nulls = [df[col].isnull().sum() for col in df.columns]
ax = sns.barplot(x = df.columns, y=nulls)
```

```python
dropped_df = df.dropna()
print("Number of records removed: " + str(df.shape[0] - dropped_df.shape[0]))
print("% of records removed: " + str(round(df.shape[0] - dropped_df.shape[0])/df.shape[0] * 100))
```

### Parsing IDR to Integers

```python
def idr_parser(cur_str):
    cur_str = re.sub("[,]", '', cur_str)
 
    if '.' in list(cur_str[-3:]):
        return cur_str[:-3]
    
    return cur_str
```

```python
converted_df = dropped_df.copy()

for col in dropped_df.columns:
    items = dropped_df[col].to_list()
    converted_df[col] = [int(idr_parser(item)) for item in items]
```

# Training Data Preparation

```python
num_features = 1 # univariate
data = converted_df.filter(['close'])
data.values
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
data.shape
```

```python
train_size = int(len(scaled_data) * .8)

train_data = scaled_data[0:train_size,:]
test_data = scaled_data [train_size:len(scaled_data), :]
```

```python
def datagen(data, step):
    features = []
    target = []
    
    for i in range(len(data)-step-1):
        features.append(data[i:i+step, 0])
        target.append(data[i+step, 0])
    
    return np.array(features), np.array(target)
```

```python
step = 60

X_train, y_train = datagen(train_data, step)
X_test, y_test = datagen(test_data, step)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], num_features)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], num_features)
X_train.shape
```

```python
real_test_data = data.values[train_size:len(data),:]
_, real_y_test = datagen(real_test_data, step)
```

# Stacked LSTM Training

```python
lstm = Sequential()
lstm.add(LSTM(128, return_sequences=True, input_shape=(step,1)))
lstm.add(LSTM(64, return_sequences=False))
lstm.add(Dense(25))
lstm.add(Dense(1))

lstm.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=[tensorflow.keras.metrics.MeanSquaredError(),
             tensorflow.keras.metrics.RootMeanSquaredError()]
)
lstm.summary()
```

```python
lstm.fit(
    X_train, y_train,
    validation_data=(X_test,y_test),
    batch_size=1, 
    epochs=10
)
```

# Predictions and Results

```python
preds = lstm.predict(X_test)
preds = scaler.inverse_transform(preds)
rmse = math.sqrt(mean_squared_error(real_y_test, preds))
print("RMSE: ", rmse)
```

```python
train_predict=lstm.predict(X_train)
test_predict=lstm.predict(X_test)
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
plt.figure(figsize=(20,5))

plt.plot(data.values)

trainPredictPlot = np.empty_like(scaled_data)
trainPredictPlot[:,:] = np.nan
trainPredictPlot[step:len(train_predict)+step, :] = train_predict
plt.plot(trainPredictPlot)

testPredictPlot = np.empty_like(scaled_data)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(train_predict)+(step*2)+1:len(scaled_data)-1, :] = test_predict
plt.plot(testPredictPlot)

plt.legend(['Actual','Train','Test'], loc='upper left')

plt.show()
```
